```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Definimos la funciÃ³n g(x)
def g(x):
    return math.pi + 0.5 * math.sin(x / 2)

# funcion de punto fijo bÃ¡sico
def punto_fijo(x0, tol, max_iter):
    valores = [x0]
    for i in range(max_iter):
        x1 = g(x0)
        valores.append(x1)
        if abs(x1 - x0) < tol:
            break
        x0 = x1
    return valores

# ParÃ¡metros iniciales
x0 = 3  # valor inicial (puede estar entre 0 y 2Ï)
tolerancia = 0.01
max_iter = 50

# Llamamos a la funciÃ³n
resultados = punto_fijo(x0, tolerancia, max_iter)
print("Punto fijo aproximado:", round(resultados[-1], 5))
print("NÃºmero de iteraciones:", len(resultados) - 1)

# GrÃ¡fica de g(x) y y = x
x_vals = np.linspace(0, 2 * math.pi, 100)
g_vals = [math.pi + 0.5 * math.sin(x / 2) for x in x_vals]

plt.plot(x_vals, g_vals, label='g(x)', color='blue')
plt.plot(x_vals, x_vals, label='y = x', linestyle='--', color='gray')

# Dibujamos las iteraciones en la grÃ¡fica
for i in range(len(resultados) - 1):
    xi = resultados[i]
    xi1 = resultados[i+1]
    # LÃ­nea vertical
    plt.plot([xi, xi], [xi, xi1], 'ro-')
    # LÃ­nea horizontal
    plt.plot([xi, xi1], [xi1, xi1], 'ro-')

plt.title("MÃ©todo de Punto Fijo")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid(True)
plt.legend()
plt.show()
```
