```python
import math

def biseccion(f, a, b, tol, max_iter=100):
    # Verificar cambio de signo
    if f(a) * f(b) >= 0:
        return None, "La función no cambia de signo en los extremos del intervalo"

    iter_count = 0
    while (b - a) / 2 > tol and iter_count < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            return c, None
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iter_count += 1

    return (a + b) / 2, None
```

```python
import math

def secante():
    # Solicitar datos de entrada
    expr = input("Ingrese la función f(x): ")
    f = lambda x: eval(expr, {'x': x, 'math': math})
    x0 = float(input("Ingrese el primer punto inicial (x0): "))
    x1 = float(input("Ingrese el segundo punto inicial (x1): "))
    tol = float(input("Ingrese la tolerancia: "))
    max_iter = int(input("Ingrese el máximo de iteraciones: "))

    iter_count = 0
    # Iterar hasta cumplir tolerancia o máximo de iteraciones
    while iter_count < max_iter:
        if abs(f(x1)) < tol:
            return x1

        try:
            x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        except ZeroDivisionError:
            print("Error: División por cero. No se puede continuar.")
            return None

        if abs(x2 - x1) < tol:
            return x2

        x0, x1 = x1, x2
        iter_count += 1

    print("Advertencia: Máximo de iteraciones alcanzado sin converger.")
    return x1

# Ejecutar el método
resultado = secante()
if resultado is not None:
    print(f"Raíz encontrada: {resultado}")
```

```python
import math

def newton():
    # Solicitar datos de entrada
    expr_f = input("Ingrese la función f(x): ")
    f = lambda x: eval(expr_f, {'x': x, 'math': math})
    expr_df = input("Ingrese la derivada f'(x): ")
    df = lambda x: eval(expr_df, {'x': x, 'math': math})
    x0 = float(input("Ingrese el punto inicial (x0): "))
    tol = float(input("Ingrese la tolerancia: "))
    max_iter = int(input("Ingrese el máximo de iteraciones: "))

    iter_count = 0
    # Iterar hasta cumplir tolerancia o máximo de iteraciones
    while iter_count < max_iter:
        fx0 = f(x0)
        dfx0 = df(x0)

        if abs(fx0) < tol:
            return x0
        if dfx0 == 0:
            print("Error: Derivada cero. No se puede continuar.")
            return None

        x1 = x0 - fx0 / dfx0

        if abs(x1 - x0) < tol:
            return x1

        x0 = x1
        iter_count += 1

    print("Advertencia: Máximo de iteraciones alcanzado sin converger.")
    return x0

# Ejecutar el método
resultado = newton()
if resultado is not None:
    print(f"Raíz encontrada: {resultado}")
```
