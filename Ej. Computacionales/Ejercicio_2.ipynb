```python
import math

def biseccion(f, a, b, tol, max_iter=100):
    # Verificar cambio de signo
    if f(a) * f(b) >= 0:
        return None, "La función no cambia de signo en los extremos del intervalo"

    iter_count = 0
    while (b - a) / 2 > tol and iter_count < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            return c, None
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iter_count += 1

    return (a + b) / 2, None
```

```python
# Definir la función
f_a = lambda x: x - 2**(-x)

# Aplicar bisección
raiz, error = biseccion(f_a, 0, 1, 1e-5)

if error:
    print(f"Error: {error}")
else:
    print(f"Raíz encontrada en x = {raiz:.7f}")
```

```python
# Definir la función
f_b = lambda x: 2*x * math.cos(2*x) - (x + 1)**2

# Aplicar bisección en [-3, -2]
raiz1, error1 = biseccion(f_b, -3, -2, 1e-5)

if error1:
    print(f"Error en [-3,-2]: {error1}")
else:
    print(f"Raíz 1 encontrada en x = {raiz1:.7f}")
```

```python
# Aplicar bisección en [-1, 0]
raiz2, error2 = biseccion(f_b, -1, 0, 1e-5)

if error2:
    print(f"Error en [-1,0]: {error2}")
else:
    print(f"Raíz 2 encontrada en x = {raiz2:.7f}")
```
